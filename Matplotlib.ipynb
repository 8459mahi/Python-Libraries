Curve with Error Band

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,10,100)
y=np.sin(x)
error= 0.2 + 0.2 * np.sqrt(x)

plt.plot(x,y, label="sin(x)", color="blue")
plt.fill_between(x,y - error, alpha=0.3, color="blue",label="Error band")
plt.xlabel("X- axis")
plt.ylabel("Y-axis")
plt.title("Curve with Error Band")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.exp(-x / 3.0)
lower_error = 0.1 + 0.1 * np.sin(x)
upper_error = 0.2 + 0.2 * np.cos(x)

# Plot the main curve
plt.plot(x, y, label='exp(-x/3)', color='green')

# Plot asymmetric error band
plt.fill_between(x, y - lower_error, y + upper_error, color='green', alpha=0.3, label='Error Band')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Asymmetric Error Band')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def koch_snowflake(order, scale=10):
    def koch_curve(p1, p2, order):
        if order == 0:
            return [p1, p2]
        else:
            p1 = np.array(p1)
            p2 = np.array(p2)
            delta = p2 - p1
            one_third = p1 + delta / 3
            two_third = p1 + 2 * delta / 3

            # Calculate peak of the triangle
            angle = np.pi / 3  # 60 degrees
            peak = one_third + np.array([
                np.cos(angle) * (two_third[0] - one_third[0]) - np.sin(angle) * (two_third[1] - one_third[1]),
                np.sin(angle) * (two_third[0] - one_third[0]) + np.cos(angle) * (two_third[1] - one_third[1])
            ])

            return (
                koch_curve(p1, one_third, order - 1)[:-1] +
                koch_curve(one_third, peak, order - 1)[:-1] +
                koch_curve(peak, two_third, order - 1)[:-1] +
                koch_curve(two_third, p2, order - 1)
            )

    # Define initial triangle
    p1 = (0, 0)
    p2 = (scale, 0)
    height = np.sqrt(3) / 2 * scale
    p3 = (scale / 2, height)

    # Get points for all 3 sides
    side1 = koch_curve(p1, p2, order)
    side2 = koch_curve(p2, p3, order)
    side3 = koch_curve(p3, p1, order)

    return side1 + side2 + side3 + [p1]

# Create and plot the snowflake
order = 5  # Increase this for more detail
points = koch_snowflake(order)

x, y = zip(*points)
plt.figure(figsize=(8, 8))
plt.plot(x, y, color='blue')
plt.title(f"Koch Snowflake (Order {order})")
plt.axis('equal')
plt.axis('off')
plt.show()
```

```python
x, y = koch_snowflake(order=5)

plt.figure(figsize=(8, 8))
plt.axis('equal')
plt.fill(x, y)
plt.show()
```
