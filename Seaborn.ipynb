```python
import seaborn as sns
```

```python
import pandas as pd
dataset= pd.read_csv("/content/Details.csv")
```

```python
dataset.head()
```

Scatter plot

```python
sns.relplot(x="Amount",y="Profit",hue="Quantity",data=dataset)
```

Line plot

```python
sns.relplot(x="Amount",y="Profit",hue="Quantity",data=dataset,kind="line",style="Category")
```

Multiple statical relationship

```python
sns.relplot(x="Amount",y="Profit",hue="Quantity",data=dataset,col="Category")
```

Histogram distribution

```python
sns.displot(dataset,x="Quantity",binwidth=0.8,hue="Category")
```

```python
sns.displot(dataset,x="Quantity",binwidth=0.8,hue="Category")
```

```python
sns.displot(dataset, x="Quantity", col="Category")
```

```python
sns.displot(dataset, x="Quantity",y="Amount",hue="Category")
```

#Joint kernel density estimate



```python
sns.displot(dataset, x="Quantity",y="Amount",kind="kde",hue="Category")
```

Pairplot

```python
sns.pairplot(dataset)
```

Scatter

```python
sns.catplot(x="Quantity",y="Amount",data=dataset,hue="Category")
```

Box plot

```python
sns.catplot(x="Quantity",y="Amount",kind="box",data=dataset,hue="Category")
```

regression

```python
sns.regplot(x="Quantity",y="Amount",data=dataset)
```

linear model

```python
sns.lmplot(x="Quantity",y="Amount",data=dataset,hue="Category",palette="Set1");
```

Using SNS theme in other graph

```python
import numpy as np
import matplotlib.pyplot as plt
def sinplot(flip=1):
  x=np.linspace(0,14,100)
  for i in range(1,10):
    plt.plot(x,np.sin(x+i* .5) * (7 - i) * flip)
sns.set_context("paper")
sinplot()
```
